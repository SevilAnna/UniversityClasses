## **My Code:**

```python
import random
import numpy as np
```

```python
ladder_bottom = (8,21,43,50,54,62,66,80)
ladder_top =   (26,82,77,91,93,96,87,100)

snake_head = (98,95,92,83,73,69,64,59,55,52,48,46,44)
snake_tail = (28,24,51,19, 1,33,36,17, 7,11, 9, 5,22)
```

```python
simu_num = 1000000                # DECREASE BEFORE RE-RUNNING!
total_counter = [0]*simu_num
for n in range(simu_num):
  won = False
  tile = 1
  counter = 0
  while won == False:
    roll=random.randint(1,6)
    counter += 1  
    #print("turn",counter,", rolled:",roll)
    tile += roll
    for t in ladder_bottom:
      if tile == t:
        tile = ladder_top[ladder_bottom.index(t)]
        #print("ladder")
    for tt in snake_head:
      if tile == tt:
        tile = snake_tail[snake_head.index(tt)]
        #print("snake")
    if tile >= 100:
      won = True
      #print("WON!")
      tile = 100
    #print("current tile:",tile)
  #print("# of turns it took:",counter)
  total_counter[n] = counter
#print(total_counter)
print("The average number of turns is:",np.mean(total_counter))
```
